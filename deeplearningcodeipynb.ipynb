```python
# ============================
# 🧠 STEP 1: IMPORT LIBRARIES
# ============================
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# ============================
# 🧩 STEP 2: LOAD DATASET
# ============================
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

print("✅ Dataset loaded successfully!")
print("Training images shape:", x_train.shape)
print("Training labels shape:", y_train.shape)
print("Testing images shape:", x_test.shape)
print("Testing labels shape:", y_test.shape)
```

```python
# Show first 10 sample images
plt.figure(figsize=(10,2))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(x_train[i], cmap='grey')
    plt.title(y_train[i])
    plt.axis('on')
plt.show()
```

```python
# ============================
# 👀 STEP 3: VIEW SAMPLE IMAGES
# ============================

# Show one image
plt.imshow(x_train[4], cmap='gray')
plt.title(f"Label: {y_train[4]}")
plt.axis('on')
plt.show()

```

```python
# ============================
# 🔢 STEP 4: NORMALIZE DATA
# ============================
# Convert pixel values (0–255) into 0–1 range
x_train = x_train.reshape(60000, 784).astype("float32") / 255
x_test = x_test.reshape(10000, 784).astype("float32") / 255

```

```python
# ============================
# 🏗️ STEP 5: BUILD MODEL
# ============================
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),  # hidden layer
    layers.Dense(10, activation='softmax')  # output layer (10 digits)
])
```

```python

# ============================
# ⚙️ STEP 6: COMPILE MODEL
# ============================
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python

# ============================
# 🏋️ STEP 7: TRAIN MODEL
# ============================
print("\n🚀 Training started...")
model.fit(x_train, y_train, epochs=3)
```

```python

# ============================
# 🧪 STEP 8: TEST MODEL
# ============================
test_loss, test_acc = model.evaluate(x_test, y_test)
print("\n✅ Test Accuracy:", test_acc)
```

```python
# ============================
# 🧮 STEP 9: PREDICT SAMPLE IMAGE
# ============================
predictions = model.predict(x_test[:5])

print("\n🎯 Predicted Labels:", tf.argmax(predictions, axis=1).numpy())
print("🧾 Actual Labels:   ", y_test[:5])

```

```python
# ============================
# 👁️ STEP 10: SHOW PREDICTED IMAGES
# ============================
plt.figure(figsize=(10,2))
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(x_test[i].reshape(28,28), cmap='gray')
    plt.title(f"Pred: {tf.argmax(predictions[i])}\nActual: {y_test[i]}")
    plt.axis('off')
plt.show()
```
